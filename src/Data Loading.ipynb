```
""""
Utility for converting JSON-formatted review data into tabular form.
"""

file_name = "/content/drive/My Drive/text-review-generation-data2040/data/Clothing_Shoes_and_Jewelry_5.json"
col_names = ["text", "rating"]
tsv_name = file_name.replace("json", "tsv")
output = open(tsv_name, 'w')
output.write("\t".join(["rating", "text"]))
output.close()

i = 0
with open(file_name) as f:
    for line in tqdm(f, desc="Reading from JSON"):
        # do stuff with this one line
        try:
            review = json.loads(line)
        
            text = review['reviewText']
            text = ' '.join(text.split())
            rating = review["overall"]
        except Error as e:
            continue
        output = open(tsv_name, 'a')
        output.write("\n")
        output.write("\t".join([str(rating), text]))
        output.close()
        i += 1
        # for smaller sets, use:
#         if i >= 10:
#             break
```
